```python
texts = []
for document in documents:
    text = []
    doc = nlp(document)
    for w in doc:
        if not w.is_stop and not w.is_punct and not w.like_num:
            text.append(w.lemma_) # Appends the lemmatized form of the word to the text list. Lemmatization involves reducing words to their base or root form.
    texts.append(text)
#texts is a mini-corpus specifically for toxic algal bloom
print(texts)
```

```python
#creating a BOW representation of the mini-corpus
dictionary = corpora.Dictionary(texts)
print(dictionary.token2id)
```

Using the doc2bow method, which helps convert our document to bag-of-words.

```python
corpus = [dictionary.doc2bow(text) for text in texts]
corpus
```
